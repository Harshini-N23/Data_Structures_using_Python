```python
class Node:
    def __init__(self,data):
        self.data =data
        self.next =None
        
class StackADT:
    
    def __init__(self):
        self.head =None
        self.size=0
        
    def isempty(self):
        if self.head==None:
            return True
        else:
            return False
        
    def push(self,data):
        if self.head==None:
            self.head=Node(data)
        else:
            newnode =Node(data)
            newnode.next =self.head
            self.head=newnode
        self.size+=1
            
    def pop(self):
        if self.isempty():
            print("Empty Stack")
            return 
        else:
            dnode =self.head
            self.head=self.head.next
            dnode.next=None
            self.size-=1
            return dnode.data
            
            
        
    def top(self):
        if self.isempty():
            return None
        else:
            return self.head.data
        
    def display(self):
         if self.isempty():
            print("Empty Stack!!")
         else:
            temp =self.head
            while temp:
                print(temp.data, end = "-->")
                temp =temp.next
            print("\u03A6")
        
    def __len__(self):
        return self.size
```

```python
from tabulate import tabulate 
def main():
   
    S = StackADT()
    while True:
        table =  [["Stack Operations"] ,["1.Push"],["2.Pop"],["3.Display"],["4.Top"],["5.Check Empty Stack"],["6.Size of Stack"],["7.Exit"]]
        print(tabulate(table, tablefmt ="pretty", headers= "firstrow"))
        try:
            choice  =  int(input("\nChoose the operation you want to perform: "))
            if choice ==1:
                item = input("\nEnter the elements to be added: ")
                S.push(item)
            elif choice ==2:
                print("***********************************************")
                print("The element popped:", S.pop())
                print(S.display())
                print("***********************************************")
            elif choice ==3:
                print("***********************************************")
                S.display()
                print("***********************************************")
            elif choice ==4:
                print("***********************************************")
                print("Head Node:",S.top())
                print("***********************************************")
            elif choice ==5:
                print("***********************************************")
                print("Stack Empty? ",S.isempty())
                print("***********************************************")
            elif choice ==6:
                print("***********************************************")
                print("The Size of Stack:", len(S))
                print("***********************************************")
            else:
                print("Thank You!!")
                break
        except ValueError:
            print("Invalid Input")
            break
```

```python
main()
```
