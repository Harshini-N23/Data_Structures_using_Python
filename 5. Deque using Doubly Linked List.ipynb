```python
class Node:
    def __init__(self,element,prev,next):
        self.element=element
        self.prev=prev
        self.next =next

class DoublyLinkedList(Node):       
    def __init__(self):
        self.header =Node(None,None,None)
        self.trailer =Node(None,None,None)
        self.header.next =self.trailer
        self.trailer.prev =self.header
        self.size = 0

    def __len__(self):
        return self.size

    def isempty(self):
        return self.size==0

    def insert(self, e, predecessor,successor):
        newest =Node(e,predecessor,successor)
        predecessor.next =newest
        successor.prev =newest
        self.size+=1
        return newest

    def delete_node(self,node):
        predecessor = node.prev
        successor = node.next
        predecessor.next =successor
        successor.prev =predecessor
        self.size-=1
        ele=node.element
        #node.element=None
        return ele

class DequeLinked(DoublyLinkedList):
    def first(self):
        if self.isempty():
            print("Deque is Empty!")
        return self.header.next.element

    def last(self):
        if self.isempty():
            print("Deque is Empty!")
        return self.trailer.prev.element

    def insert_first(self,e):
        self.insert(e,self.header,self.header.next)

    def insert_last(self,e):
        self.insert(e,self.trailer.prev, self.trailer)

    def delete_first(self):
        if self.isempty():
            print("Deque is Empty!")
        else:
            ele1 = self.delete_node(self.header.next)
            print("The element deleted: ", ele1)
            
    def delete_last(self):
        if self.isempty():
            print("Deque is Empty!")
        else:
            ele2 = self.delete_node(self.trailer.prev)
            print("The element deleted: ", ele2)
            
    def display(self):
            if self.isempty():
                print("Empty Deque!")
            else:
                temp =self.header
                while temp:
                    print(temp.element, end="  ")
                    temp =temp.next
                print()
```

```python
from tabulate import tabulate
def main():
    D = DequeLinked()
    while True:
        table = [["Deque - Doubly Linked List"] ,["1.Insert at beginning"],["2.Insert at the end"],["3.Display"],["4.Size"],["5.IsEmpty"],["6.Delete at First"],["7.Delete at the End"],["8.Exit"]]
        print(tabulate(table, tablefmt ="pretty", headers= "firstrow"))
        try:
            choice = int(input("\nChoose the operation you want to perform: "))
            if choice ==1:
                front = input("\nEnter the elements to be added: ")
                D.insert_first(front)
                D.display()
            elif choice ==2:
                item = input("\nEnter the elements to be added: ")
                D.insert_last(item)
                D.display()
            elif choice==3:
                D.display()
            elif choice==4:
                 print("Size of Linked List: ", len(D))
            elif choice==5:
                 print("Linked List Empty?", D.isempty())
            elif choice==6:
                D.delete_first()
            elif choice==7:
                D.delete_last()
            else:
                print("Thank You")
                break
        except ValueError:
            print("Invalid Input")
```
