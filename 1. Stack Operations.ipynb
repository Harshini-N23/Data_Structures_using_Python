# Using List

```python
class Stack:
    def __init__(self):
        self.values =[]
    
    def push(self, value):
        if len(self.values)>100:
            print("Stack Overflow")
        else:
            return self.values.append(value)
    
    def pop(self):
        if len(self.values)==0:
            print("Empty stack!!")
        else:
            print("The element popped: ", self.values.pop())
            print("and the stack consists of: ",self.values)
    
    def display(self):
        print("The elements in Stack: ",self.values)
    
    def top(self):
        if len(self.values)==0:
            print("Empty Stack!!")
        else:
            print("The first element of Stack: ",self.values[-1])
    
    def empty(self):
        print("Stack Empty?", len(self.values)==0)
    
    def size(self):
        print("The number of elements in stack is:", len(self.values))
```

```python
from tabulate import tabulate 
def main():
   
    S = Stack()
    while True:
        table =  [["Stack Operations"] ,["1.Push"],["2.Pop"],["3.Display"],["4.Top"],["5.Check Empty Stack"],["6.Size of Stack"],["7.Exit"]]
        print(tabulate(table, tablefmt ="pretty", headers= "firstrow"))
        try:

            choice  =  int(input("\nChoose the operation you want to perform: "))
            if choice ==1:
                item = input("\nEnter the elements to be added: ")
                S.push(item)
            elif choice ==2:
                print("***********************************************")
                S.pop()
                print("***********************************************")
            elif choice ==3:
                print("***********************************************")
                S.display()
                print("***********************************************")
            elif choice ==4:
                print("***********************************************")
                S.top()
                print("***********************************************")
            elif choice ==5:
                print("***********************************************")
                S.empty()
                print("***********************************************")
            elif choice ==6:
                print("***********************************************")
                S.size()
                print("***********************************************")
            else:
                print("Thank You!!")
                break
        except ValueError:
            print("Not a valid choice!")
            break
```

```python
main()
```

# Using collections.deque

```python
from collections import deque
```

```python
class DStack:
    def __init__(self):
        self.values = deque()
    
    def push(self, value):
        if len(self.values)>2:
            print("Stack Overflow")
        else:
            return self.values.append(value)
    
    def pop(self):
        if len(self.values)==0:
            print("Trying to delete elements from empty stack! Please add elements")
        else:
            return self.values.pop()
    
    def display(self):
        return self.values
    
    def top(self):
        return self.values[-1]
    
    def empty(self):
        return len(self.values)==0
    
    def size(self):
        return len(self.values)
```

```python
from tabulate import tabulate 
def main():
    table =  [["Stack Operations"] ,["1.Push"],["2.Pop"],["3.Display"],["4.Top"],["5.Check Empty Stack"],["6.Size of Stack"]]
    print(tabulate(table, tablefmt ="pretty", headers= "firstrow"))
    DS = DStack()
    while True:
        try:

            choice  =  int(input("\nChoose the operation you want to perform: "))
            if choice ==1:
                item = input("\nEnter the elements to be added: ")
                DS.push(item)
            elif choice ==2:
                print("***********************************************")
                print("The element popped: ", DS.pop())
                print("and the stack consists of: ", DS.display())
                print("***********************************************")
            elif choice ==3:
                print("***********************************************")
                print("The elements in Stack: ",DS.display())
                print("***********************************************")
            elif choice ==4:
                if DS.empty():
                    print("Empty Stack!!")
                else:
                    print("***********************************************")
                    print("The first element of Stack: ",DS.top())
                    print("***********************************************")
            elif choice ==5:
                print("***********************************************")
                print("Stack Empty?",  DS.empty())
                print("***********************************************")
            elif choice ==6:
                print("***********************************************")
                print("The number of elements in stack is:", DS.size())
                print("***********************************************")
            else:
                print("Invalid choice")
        except ValueError:
            print("Not a valid choice!")
            break
```

```python
main()
```

# Using Queue Module

```python
from queue import LifoQueue
```

```python
class QStack:
    def __init__(self):
        self.values = LifoQueue(maxsize=2)
    
    def push(self, value):
#         if self.values.qsize()>maxsize:
#             print("Stack Overflow")
#         else:
            return self.values.put(value)
    
    def pop(self):
        if self.values.empty():
            print("Trying to delete elements from empty stack! Please add elements")
        else:
            return self.values.get()
    
    def display(self):
        return self.values
    
#     def top(self):
#        return self.values[-1]
    
    def empty(self):
        return self.values.empty()
    
    def size(self):
        return self.values.qsize()
```
