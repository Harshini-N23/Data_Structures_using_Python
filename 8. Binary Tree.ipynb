```python
class Node:
    __slots__ = "element", "left","right"
   

    def __init__(self,E, L=None, R=None):
        self.element = E
        self.left =L
        self.right=R
       
   
class BinaryTree:
    def __init__(self):
        self.root =None
       
    def maketree(self,e,left,right):
        self.root =Node(e,left.root,right.root)
       
    def inorder(self,troot):
        if troot:
            self.inorder(troot.left)
            print(troot.element, end=", ")
            self.inorder(troot.right)
    def preorder(self, troot):
        if troot:
            print(troot.element, end=", ")
            self.preorder(troot.left)
            self.preorder(troot.right)
    
    def postorder(self,troot):
        if troot:
            self.postorder(troot.left)
            self.postorder(troot.right)
            print(troot.element, end=", ")
              
    def maxDepth(self, node):
        if node is None:
            return -1
        else :
            lDepth = self.maxDepth(node.left)
            rDepth = self.maxDepth(node.right)
        if (lDepth > rDepth):
            return lDepth+1
        else:
            return rDepth+1
        
    def leaf(self, root):
        s1 = []
        s2 = []
        s1.append(root)
        while len(s1) != 0:
            curr = s1.pop()
            if curr.left:
                s1.append(curr.left)
            if curr.right:
                s1.append(curr.right)
            elif not curr.left and not curr.right:
                s2.append(curr)
        a = []       
        while len(s2) != 0:
            a.append(s2.pop().element)
        return a

    def count_nodes(self,node):
        if node is None:
            return 0
        return 1 + self.count_nodes(node.left) + self.count_nodes(node.right)


import matplotlib.pyplot as plt
import matplotlib.image as mpimg
def main():
   
    img = mpimg.imread('Plant_classification_diagram.png')
    imgplot = plt.imshow(img)
    plt.figure(figsize=(10,20))
    plt.show()
    A= BinaryTree()
    B= BinaryTree()
    C= BinaryTree()
    D= BinaryTree()
    E= BinaryTree()
    F= BinaryTree()
    G= BinaryTree()
    H= BinaryTree()
    I= BinaryTree()
    phi= BinaryTree()

    I.maketree("Grass",phi,phi)
    G.maketree("Angiosperms",I,phi)
    H.maketree("Gymnosperms",phi,phi)
    D.maketree("Seed",G,H)
    E.maketree("Spore",phi,phi)
    F.maketree("Algae", phi,phi)
    B.maketree("Vascular",D,E)
    C.maketree("Non_Vascular", F,phi)
    A.maketree("Plant_Kingdom", B,C)
    
    print("Total Nodes: ", A.count_nodes(A.root))
    print("-------------------------------------------------------------------------------------------")
    print("Inorder Traversal:",)
    A.inorder(A.root)
    print("\n-------------------------------------------------------------------------------------------")
    print("Preorder Traversal:")
    A.preorder(A.root)
    print("\n-------------------------------------------------------------------------------------------")
    print("Postorder Traversal:")
    A.postorder(A.root)
    print("\n-------------------------------------------------------------------------------------------")
    
   
    print("Height of the Tree: ", end=" ")
    print(A.maxDepth(A.root))
    print("-------------------------------------------------------------------------------------------")
  
    
    print("Leaf Nodes:")
    print(A.leaf(A.root))
    print("-------------------------------------------------------------------------------------------")
    
   
       
main()    
        
```
