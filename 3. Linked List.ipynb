```python
class Node:
    def __init__(self,data,next):
        self.data = data
        self.next = next
        
class LinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0

    def __len__(self):
        return self.size
   
    def isempty(self):
        return self.size==0
    
    def insertfront(self,new_data):
        new_node =Node(new_data,None)
        if self.isempty():
            self.head=new_node
        else:
            new_node.next = self.head
            self.head=new_node
        self.size +=1
    
    def insertposition(self,new_data,position):
        new_node = Node(new_data,None)
        if position<0:
            print("Position should be >=0.")
        elif position ==0:
            self.insertfront(new_data)
        else:
            temp2 = self.head
            for i in range(0,position-1):
                if(temp2!=None):
                    temp2=temp2.next
            if temp2!=None:
                new_node.next = temp2.next
                temp2.next=new_node
            else:
                print("Previous Node is Null")
            self.size +=1
    
    def insertend(self,data):
        new_node = Node(data, None)
        if self.head == None:
            self.head = new_node
            return 
        last = self.head
        while (last.next):
            last = last.next
        last.next =  new_node
        self.size+=1
        
    def search(self,ele):
        i = []
        temp1 = self.head
        index = 0
        if self.isempty():
            print("List is Empty!!")
        else:
            while temp1!= None:
                if temp1.data == ele:
                    i.append(index)
                if temp1.data.isupper() == ele:
                    i.append(index)
                if temp1.data.islower() ==ele:
                    i.append(index)
                temp1=temp1.next
                index=index+1
        return i
       
    def display(self):
        if self.isempty():
            print("No Linked List has been created")
        else:
            temp =self.head
            while temp:
                print(temp.data, end = "-->")
                temp =temp.next
            print("\u03A6")
            
    def delete(self,key):
        temp = self.head
        if self.isempty():
            print("List is empty!!")       
        else:           
            if (temp is not None):
                if (temp.data == key):
                    self.head = temp.next
                    temp = None
                    self.size-=1
                    return
                
        while(temp is not None):
            if temp.data == key:
                break
            prev = temp
            temp = temp.next
        if(temp == None):
            print("Element not Found")
            return
        self.size-=1
        prev.next = temp.next 
        temp = None
```

```python
from tabulate import tabulate
def main():
   
    L  = LinkedList()
    while True:
        table =  [["Linked List"] ,["1.Insert at beginning"],["2.Insert at a Position"],["3.Insert at the end"],["4.Display"],["5.Search"],["6.Size"],["7.IsEmpty"],["8.Deletion"],["9.Exit"]]
        print(tabulate(table, tablefmt ="pretty", headers= "firstrow"))
        try:
            choice  =  int(input("\nChoose the operation you want to perform: "))
            if choice ==1:
                front = input("\nEnter the elements to be added: ")
                L.insertfront(front)
                L.display()
            elif choice ==2:
                pos = int(input("Enter the position\n(*Index starts from 0): "))
                value = input("Enter the value to be inserted: ")
                L.insertposition(value,pos)
                L.display()
            elif choice==3:
                item = input("\nEnter the elements to be added: ")
                L.insertend(item)
                L.display()
            elif choice==4:
                L.display()
            elif choice==5:
                se =input("Enter the element to be searched: ")
                ind = L.search(se)
                if len(ind)==0:
                    print("Element not Found")
                else:
                    print(se, "found at index:",ind)
            elif choice==6:
                print("Size of Linked List: ", len(L))
            elif choice==7:
                print("Linked List Empty?", L.isempty())
            elif choice ==8:
                key=input("Enter the element to be deleted ")
                a=len(L.search(key))
                if a>1:
                    for i in range(a):
                        L.delete(key)
                       # L.display()
                else:
                    L.delete(key)
                L.display()
            else:
                print("Thank You")
                break
        except ValueError:
            print("Invalid Input")
            break
```

```python
main()
```
