# Using List

```python
class Queue:
    def __init__(self):
        self.values = []
        
    def enqueue(self, value):
        return self.values.append(value)
    
    def dequeue(self):
        if len(self.values)==0:
            print( "Queue is empty!")
        else:
            return self.values.pop(0)
        
    def display(self):
        return self.values
    
    def isempty(self):
        return len(self.values)==0
    
    def size(self):
        return len(self.values)
    
    def front(self):
        if self.isempty():
             print("Empty Queue!!")
        else:
            return self.values[0]
    
    def rear(self):
         if self.isempty():
            print("Empty Queue!!")
         else:
            return self.values[-1]
```

```python
from tabulate import tabulate 

def main():
        
        Q = Queue()
        while True:
            table =  [["Queue Operations"] ,["1.Enque"],["2.Dequeue"],["3.Front and Rear Element"],["4.Check Empty Queue"],["5.Size of Queue"], ["6.Exit"]]
            print(tabulate(table, tablefmt = "pretty", headers ="firstrow"))
            try:
                
                choice  =  int(input("\nChoose the operation you want to perform: "))
                if choice ==1:
                    item = input("\nEnter the elements to be added: ")
                    Q.enqueue(item) 
                    print(Q.display())
                elif choice ==2:
                    if Q.isempty():
                        print("Empty Queue")
                    else:
                        print("***************************************************")
                        print("The element popped: ", Q.dequeue())
                        print("and the queue consists of:", Q.display())
                        print("***************************************************")
                elif choice ==3:
                    if Q.isempty():
                        print("Empty Queue")
                    else:
                        print("***************************************************")
                        print("Front Element: ",Q.front(), "\nRear Element: ", Q.rear())
                        print("***************************************************")
                elif choice ==4:
                    print("***************************************************")
                    print("Queue Empty?",  Q.isempty())
                    print("***************************************************")
                elif choice ==5:
                    print("***************************************************")
                    print("The number of elements in Queue is:", Q.size())
                    print("***************************************************")
                else:
                    print("Thank You")
                    break
            except ValueError:
                print("Not a valid choice!")
                break
```

```python
main()
```

# Using deque class

```python
from collections import deque
```

```python
class Cqueue:
    def __init__(self):
        self.values = deque()
   
    def enqueue(self,value):
        return self.values.append(value)

    def dequeue(self):
        if len(self.values)==0:
            print( "Queue is empty!")
        else:
            print("The element popped",self.values.popleft())
            print("and the queue consists of: ", self.values)
       
    def display(self):
        print("The elements in the Queue: ",self.values)
        
    def isempty(self):
        print("Queue Empty? ",len(self.values)==0)
    
    def front(self):
        if len(self.values)==0:
             print("Empty Queue!!")
        else:
            print("Front Element: ",self.values[0])
            
    def size(self):
        print("The number of elements in Queue: ",len(self.values))
   
    def rear(self):
         if len(self.values)==0:
            print("Empty Queue!!")
         else:
            print("Rear Element: ",self.values[-1])



```

```python
from tabulate import tabulate

def main():
    C = Cqueue()
    while True:
        table =  [["Queue Operations"] ,["1.Enque"],["2.Dequeue"],["3.Front and Rear"],["4.Empty"],["5.Size"],["6.Exit"]]
        print(tabulate(table, tablefmt = "pretty", headers ="firstrow"))
        try:
            choice  =  int(input("\nChoose the operation you want to perform: "))
            if choice ==1:
                item = input("\nEnter the elements to be added: ")
                C.enqueue(item)
                C.display()
            elif choice ==2:
                    print("***************************************************")
                    C.dequeue()
                    print("***************************************************")
            elif choice ==3:
                    print("***************************************************")
                    C.front()
                    C.rear()
                    print("***************************************************")
            elif choice ==4:
                print("***************************************************")
                C.isempty()
                print("***************************************************")      
            elif choice ==5:
                print("***************************************************")
                C.size()
                print("***************************************************")
            else:
                print("Thank You!!")
                break
        except ValueError:
            print("Not a valid choice!")
            break
```

```python
main()
```
