```python
def BubbleSort(li):
    n = len(li)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if li[j] > li[j+1] :
                li[j], li[j+1] = li[j+1], li[j]
                swapped = True
        if swapped == False:
            break
    return li

def SelectionSort(lis):
    for i in range(len(lis)):
        min_idx = i
        for j in range(i+1, len(lis)):
            if lis[min_idx] > lis[j]:
                min_idx = j    
        lis[i], lis[min_idx] = lis[min_idx], lis[i]
    return lis
       
import time
import random
import string
def main():
    while True:
        try:
            list1 = []
            num = int(input("\n**SORTING TECHNIQUES**\nEnter the number of elements in the list: "))
            if num<=1:
                print("Number of elements should be greater than 1")
                break
            #print("Enter the elements: ")
            for i in range(num):
                #ele  = random.randint(-100,100)
                ele  = random.choices(string.ascii_letters+string.digits+string.punctuation)
                list1.append(ele)
            print("The generated random numbers are: ", list1)
            print("\n*Sort Method*\n1.Bubble Sort\n2.Selection Sort\n3.Exit")
            choice = int(input("\nEnter the Sorting Method: "))
            if choice ==1:
                print("---------------------------------------------------------------------------------------------")
                start = time.time()
                print("The elements present in the list are: ", list1)
                BS = BubbleSort(list1)
                print("The Sorted list: ", BS)
                print("Descending Order: ", sorted(BS, reverse=True))
                end =time.time()
                print("Execution Time: ", end-start)
                print(list1)
                print("---------------------------------------------------------------------------------------------")
            elif choice ==2:
                print("---------------------------------------------------------------------------------------------")
                strt =time.time()
                print("The elements present in the list are: ", list1)
                IS = SelectionSort(list1)
                print("The Sorted list: ", IS)
                en = time.time()
                print("Execution Time: ", en-strt)
                print("---------------------------------------------------------------------------------------------")
            else:
                print("Thank You")
                break
   
        except ValueError:
            print("Invalid Input")
            break



main()            
```
